```python
import numpy as np
import matplotlib.pyplot as plt
import ode
```

# Modeling a speaker driver as a simple harmonic oscillator

## A few assumptions

-First we will assume that the z- and y-components of force cancel out since the driver is circular in shape and all "springs" will be acting evenly in either direction $$\vec{F}_{nety} = \vec{F}_{netz} = 0$$

-Second we will attach our coil, spider, and surround by fixed "poles" such that there are now three "springs" working in parallel to create an overall force on the cone with equal stretch values

-Third we will set the mass of the cone to 1kg so that we can focus entirely on the effect of different force functions

```python
#define functions
def springforce(y0, t):
    dydt = np.zeros(2)
    dydt[0] = y0[1]
    if np.abs(y0[0])> 1.5:
        dydt[1] = (-ksp*(y0[0]))+(-ksu*(y0[0]+y0[0]**3))+(-kcoil*(y0[0]))
    else:
        dydt[1] = (-ksp*(y0[0]))+(-ksu*(y0[0]))+(-kcoil*(y0[0]))
    return dydt

def lf_spring(id, x, v, t):
    if(id==0): #id=0 returns dx/dt
        return v
    elif np.abs(x)>1.5:
        return (-ksp*(y0[0]+y0[0]))+(-ksu*(y0[0]+y0[0]**3))+(-kcoil*(y0[0]))
    else:
        return (-ksp*(y0[0]))+(-ksu*(y0[0]))+(-kcoil*(y0[0]))
    
def lf_springlin(id, x, v, t):
    if(id==0): #id=0 returns dx/dt
        return v
    else:
        return (-ksp*(y0[2]))+(-ksu*(y0[2]))+(-kcoil*(y0[2]))
```

The springforce differential equations come from a sum of the three spring forces, $$\vec{F}_{net}= -k_1x +(-k_2x)+(-k_3x)$$

When the spring gets close to the maximum displacement the force becomes nonlinear and can be described as $$\vec{F}_{net}= -k_1(x+x) +(-k_2(x+x^3))+(-k_3x)$$

where x is the stretch, $k_1$ is the stiffness of the spider, $k_2$ is the stiffness of the surround, and $k_3$ is the stiffness of the coil

```python
ksp = 50 #stiffness of the spider
ksu = 100 #stiffness of the surround
kcoil = 5 #stiffness of the coil


#Vsp0 = 5 #initial x-velocity of the spider
#Vsu0 = 5 #initial x-velocity of the surround
Vcoil0 = 5 #initial x-velocity of the coil

#rsp0 = 3 #initial x-position of the spider
#rsu0 = 6 #initial x-position of the surround
rcoil0 = -3 #initial x-position of the coil

y0 = [rcoil0, Vcoil0, rcoil0, Vcoil0] #initial x position, initial x velocity for each spring
t = 0.0
h = 0.0001
ta, rcoil, Vcoil, Fcoil, rcoillin, Vcoillin, Fcoillin = [],[],[],[],[],[],[] #lists to store time, and numerical x-position for each spring

while t<1.0:
    ta.append(t)
    rcoil.append(y0[0]) #numerical nonlinear coil position
    Vcoil.append(y0[1]) #numerical nonlinear coil velocity
    Fcoil.append((-ksp*(y0[0]+y0[0]))+(-ksu*(y0[0]+y0[0]**3))+(-kcoil*(y0[0]))) #numerical total force on the nonlinear coil
    rcoillin.append(y0[2]) #numerical linear coil position
    Vcoillin.append(y0[3]) #numerical linear coil velocity
    Fcoillin.append((-ksp*(y0[2]))+(-ksu*(y0[2]))+(-kcoil*(y0[2]))) #numerical total force on the linear coil
    
    ####### Solver #######
    #RK4
    #y1 = RK4(springforce, y0, t, h) #update y[n]

    #Leapfrog
    y1 = ode.leapfrog(lf_spring, y0[0], y0[1], t, h)
    y2 = ode.leapfrog(lf_springlin, y0[2], y0[3], t, h)
    
    for i in range(len(y1)): #reset y0 to the new vector
        y0[i] = y1[i]

    for i in range(len(y2)): #reset y0 to the new vector
        y0[i+2] = y2[i]    
    
    t = t + h #update clock

    
plt.figure()
plt.plot(ta,rcoil, 'g-', label='nonlinear coil position')
plt.plot(ta,rcoillin, 'b--', label='linear coil position')
plt.xlabel('t (s)')
plt.ylabel('y (m)')
plt.legend()
plt.show()

plt.figure()
plt.plot(rcoil,Vcoil, 'g-', label='nonlinear coil velocity')
plt.plot(rcoillin,Vcoillin, 'b--', label='linear coil velocity')
plt.xlabel('x (m)')
plt.ylabel('V (m/s)')
plt.legend()
plt.show()

plt.figure()
plt.plot(rcoil,Fcoil, 'g-', label='Force nonlinear springs')
plt.plot(rcoillin,Fcoillin, 'b--', label='Force linear springs')
plt.xlabel('x (m)')
plt.ylabel('F (N)')
plt.legend()
plt.show()
```

#### The graphs above show the x-displacement versus time, the velocity versus x-displacement, and the force versus x-displacement

The model of Force versus time was based on the graph given in Loudpeaker Nonlinearalities, the other two graphs are to be able to compare the ideal(linear) springs to the un-ideal(nonlinear) springs. With the same starting parameters it is clear that the nonlinear springs will have a greater frequncy as well as larger maximum velocity by a significant amount. It would be interesting to experiment and see if these differences account for any signal distortion in a classical loudspeaker.

With the three springs working in parallel the function graphed looks identical to a simple harmonic oscillator. However you can see that the force due to the springs is non-linear as the spring displacement pushes outside of 1.5m the spring force becomes exponential.

the source for comparison of the model is [here](https://www.klippel.de/fileadmin/_migrated/content_uploads/Loudspeaker_Nonlinearities%E2%80%93Causes_Parameters_Symptoms_01.pdf)
